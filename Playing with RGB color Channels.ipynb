## Removing the blue component in RGB

```python
import cv2
img=cv2.imread("flower.png")

img[:,:,0]=0

cv2.imshow("window",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img=cv2.imread("Boy.png")

imgBlue=img[:,:,0]
imgGreen=img[:,:,1]
imgRed=img[:,:,2]
new_img=np.hstack((imgBlue,imgGreen,imgRed))

cv2.imshow("window",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
