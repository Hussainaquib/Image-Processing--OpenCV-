## Vertical axis flip

```python
import cv2
img=cv2.imread("Boy.png")

img_flip=cv2.flip(img,0)

cv2.imshow("window",img_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Horizontal axis flip

```python
import cv2
img=cv2.imread("Boy.png")

img_flip=cv2.flip(img,1)

cv2.imshow("window",img_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Combine (Horizontal and vertical axis) flip

```python
import cv2
img=cv2.imread("Boy.png")

img_flip=cv2.flip(img,-1)

cv2.imshow("window",img_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
