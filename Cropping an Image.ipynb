## Crop Image using numpy
In image left top corner is origin

```python
import cv2
img=cv2.imread("Boy.png")

img_crop=img[100:300,200:500]

cv2.imshow("window",img_crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread("Boy.png")

img_crop=img[0:300,0:300]

cv2.imshow("window",img_crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
